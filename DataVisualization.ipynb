# Pie chart

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels)
plt.show()
```

```python
fig2, ax2 = plt.subplots()
ax2.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
ax2.axis('equal')
plt.show()
```

How if I want to do it one by one?

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.barh([0], df['Go-Ride'].iloc[0], 0.1, tick_label=['Go-Ride'])
ax.barh([0], df['Go-Ride'].iloc[1], 0.1, tick_label=['Go-Ride'], left=df['Go-Ride'].iloc[0], color='y')
ax.tick_params(top='off', bottom='off', left='on', right='off', labelleft='on', labelbottom='off')
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.barh([0], df['Go-Ride'].iloc[0], 0.1, tick_label=['Go-Ride'])
ax.barh([0], df['Go-Ride'].iloc[1], 0.1, tick_label=['Go-Ride'], left=df['Go-Ride'].iloc[0], color='y')
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

fig, ax = plt.subplots(figsize=(10,2))
ax.barh([0], df['Go-Food'].iloc[0], 0.1, tick_label=['Go-Food'])
ax.barh([0], df['Go-Food'].iloc[1], 0.1, tick_label=['Go-Food'], left=df['Go-Food'].iloc[0], color='y')
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
```

```python
for column in df.columns:
    fig, ax = plt.subplots(figsize=(10,1))
    ax.barh([0], df[column].iloc[0], 0.1, tick_label=[column])
    ax.barh([0], df[column].iloc[1], 0.1, tick_label=[column], left=df[column].iloc[0], color='y')
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
```

```python
for column in df.columns:
    fig, ax = plt.subplots(figsize=(10,1))
    ax.barh([0], df[column].iloc[0], 0.1)
    ax.barh([0], df[column].iloc[1], 0.1, left=df[column].iloc[0], color='y')
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.tick_params(top='off', bottom='on', left='off', right='off', labelleft='off', labelbottom='off')
    ax.set_title(column)
```

How if I want to make it in one figure but doing it separately like before?

https://matplotlib.org/examples/pylab_examples/subplots_demo.html

```python
fig, ax = plt.subplots(2, sharex=True, sharey=True, figsize=(10,1))
for i in range(len(df.columns)):
    column = df.columns[i]
    #fig, axarr[i] = plt.subplots(figsize=(10,1))
    ax[i].barh([0], df[column].iloc[0], 0.1)
    ax[i].barh([0], df[column].iloc[1], 0.1, left=df[column].iloc[0], color='y')
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['bottom'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['left'].set_visible(False)
    ax[i].tick_params(top='off', bottom='on', left='off', right='off', labelleft='off', labelbottom='on')
    ax[i].set_title(column, x=0.05,y=0)
```
